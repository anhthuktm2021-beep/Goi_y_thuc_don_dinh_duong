```python
import pandas as pd
df = pd.read_csv('nutrients_csvfile.csv')
display(df.head())
```

```python
# Xem tổng quan các cột và kiểu dữ liệu
df.info()

# Đếm số lượng giá trị bị trống (NaN - Not a Number) trong từng cột
print(df.isnull().sum())
```

```python
# Đổi tên về tiếng Anh cho đồng bộ với dữ liệu gốc
df = df.rename(columns={
    'Food': 'Food_Name',
    'Grams': 'Grams',
    'Calories': 'Calories',
    'Protein': 'Protein',
    'Fat': 'Fat',
    'Carbs': 'Carbs',
    'Category': 'Category'
})
```

```python
# Cách 1: Xóa luôn những dòng bị thiếu thông tin dinh dưỡng
df = df.dropna()

# Loại bỏ các món ăn bị trùng lặp tên (nếu có)
df = df.drop_duplicates(subset=['Food_Name'])
```

```python
# Danh sách cột số mới (đã đổi tên sang tiếng Anh)
cac_cot_so = ['Grams', 'Calories', 'Protein', 'Fat', 'Carbs']

for cot in cac_cot_so:
    df[cot] = df[cot].astype(str).str.replace(',', '').str.replace('t', '0')
    df[cot] = pd.to_numeric(df[cot], errors='coerce').fillna(0)

display(df.head())
```

```python
# Lưu thành file mới, index=False để không lưu thêm cột số thứ tự

df.to_csv('cleaned_nutrition.csv', index=False)

print("Đã lưu dữ liệu làm sạch thành công!")
```
