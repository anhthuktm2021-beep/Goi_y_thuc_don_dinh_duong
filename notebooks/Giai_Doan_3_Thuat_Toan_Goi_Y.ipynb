```python
import pandas as pd
import random

#Đọc dữ liệu sạch từ GitHub
url = 'https://raw.githubusercontent.com/anhthuktm2021-beep/Goi_y_thuc_don_dinh_duong/refs/heads/main/cleaned_nutrition.csv'
df = pd.read_csv(url)

target_calories = tdee_target

print(f"Mục tiêu: {target_calories} Kcal | Protein: {target_protein}g | Carbs: {target_carbs}g | Fat: {target_fat}g")
print(f"Tổng số món ăn trong hệ thống: {len(df)} món")
```

```python
# Tạo ra các danh sách thức ăn theo thế mạnh dinh dưỡng
# Ví dụ: Món nào có lượng Protein lớn hơn 15g/100g thì cho vào nhóm Giàu Đạm
giau_dam = df[df['Protein'] >= 15]
giau_tinh_bot = df[df['Carbs'] >= 30]
giau_beo = df[df['Fat'] >= 15]

print(f"Số món giàu Đạm: {len(giau_dam)}")
print(f"Số món giàu Tinh bột: {len(giau_tinh_bot)}")
print(f"Số món giàu Chất béo: {len(giau_beo)}")
```

```python
# Hàm chia nhỏ mục tiêu
def chia_bua_an(target, p, c, f):
    bua_sang = {'cal': target * 0.25, 'p': p * 0.25, 'c': c * 0.25, 'f': f * 0.25}
    bua_trua = {'cal': target * 0.40, 'p': p * 0.40, 'c': c * 0.40, 'f': f * 0.40}
    bua_chieu = {'cal': target * 0.35, 'p': p * 0.35, 'c': c * 0.35, 'f': f * 0.35}
    return bua_sang, bua_trua, bua_chieu

sang, trua, chieu = chia_bua_an(tdee_target, target_protein, target_carbs, target_fat)
```

```python
import random

def goi_y_bua_an(df_data, target_dict, ten_bua_an):
    # Lọc danh sách món ăn phù hợp với calo của bữa đó trước để tránh bốc món quá to
    # Ví dụ: bữa sáng 500kcal thì không thể bốc món 1000kcal
    danh_sach_phu_hop = df_data[df_data['Calories'] <= target_dict['cal']]

    # Vòng lặp tìm kiếm: Thử bốc 2 món kết hợp lại
    for i in range(1000): # Thử tối đa 1000 lần
        mon_1 = danh_sach_phu_hop.sample(n=1).iloc[0]
        mon_2 = danh_sach_phu_hop.sample(n=1).iloc[0]

        tong_cal = mon_1['Calories'] + mon_2['Calories']
        tong_p = mon_1['Protein'] + mon_2['Protein']

        # Kiểm tra điều kiện: Sai số Calo và Protein dưới 15% là chấp nhận được
        if abs(tong_cal - target_dict['cal']) <= target_dict['cal'] * 0.15:
            print(f"--- {ten_bua_an} ---")
            print(f"1. {mon_1['Food_Name']} ({mon_1['Grams']}g)")
            print(f"2. {mon_2['Food_Name']} ({mon_2['Grams']}g)")
            print(f"Tổng Calo bữa này: {tong_cal:.1f} kcal (Mục tiêu: {target_dict['cal']:.1f})")
            return tong_cal, tong_p

    print(f"Không tìm thấy tổ hợp phù hợp cho {ten_bua_an}")
    return 0, 0

# Chạy thử thuật toán cho cả ngày
print("GỢI Ý THỰC ĐƠN CỦA BẠNG HÔM NAY:\n")
c1, p1 = goi_y_bua_an(df, sang, "BỮA SÁNG")
c2, p2 = goi_y_bua_an(df, trua, "BỮA TRƯA")
c3, p3 = goi_y_bua_an(df, chieu, "BỮA CHIỀU")

print("\n" + "="*30)
print(f"TỔNG CỘNG CẢ NGÀY: {c1+c2+c3:.1f} / {tdee_target} kcal")
```
