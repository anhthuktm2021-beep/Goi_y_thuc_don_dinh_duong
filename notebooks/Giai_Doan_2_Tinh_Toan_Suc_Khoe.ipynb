```python
def tinh_toan_suc_khoe():
    print("--- CHƯƠNG TRÌNH PHÂN TÍCH SỨC KHỎE CÁ NHÂN ---")

    # 1. Nhập liệu từ người dùng
    gender = input("Giới tính của bạng (Nam/Nu): ").strip().lower()
    weight = float(input("Nhập cân nặng (kg): "))
    height = float(input("Nhập chiều cao (cm): "))
    age = int(input("Nhập tuổi: "))

    print("\nChọn mức độ vận động của bạng:")
    print("1. Ít vận động (Chỉ ăn, ngủ, học bài)")
    print("2. Vận động nhẹ (Tập thể dục 1-3 buổi/tuần)")
    print("3. Vận động vừa (Tập thể dục 3-5 buổi/tuần)")
    print("4. Vận động nặng (Tập thể dục 6-7 buổi/tuần)")
    active_level = int(input("Nhập số tương ứng (1-4): "))

    # 2. Tính BMR
    if gender == 'nam':
        s = 5
    else:
        s = -161

    bmr = 10 * weight + 6.25 * height - 5 * age + s

    # 3. Tính TDEE dựa trên mức độ vận động
    r_map = {1: 1.2, 2: 1.375, 3: 1.55, 4: 1.725}
    r = r_map.get(active_level, 1.2) # Mặc định là 1.2 nếu nhập sai

    tdee = bmr * r

    print("-" * 30)
    print(f"Chỉ số BMR của bạng: {bmr:.2f} kcal")
    print(f"Tổng năng lượng bạng tiêu thụ mỗi ngày (TDEE): {tdee:.2f} kcal")
    print("-" * 30)

    return tdee

# Chạy thử hàm
tdee_nguoi_dung = tinh_toan_suc_khoe()
```
